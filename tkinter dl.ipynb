```python
import tkinter
from tkinter import messagebox
from tkinter import filedialog
import pandas as pd
import numpy as np
import os
import math
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
```

```python
def download_series_2(series_list, start_date, end_date = None):
    driver_5 = webdriver.Chrome()
    for series in series_list:
        driver_5.get(str("https://fred.stlouisfed.org/series/" + str(series)))
        element = WebDriverWait(driver_5, 20).until(EC.presence_of_element_located((By.XPATH, "//div[@class='highcharts-legend']")))
        driver_5.find_element_by_xpath("//span[@id='zoom-all']").click()
        start_input = driver_5.find_element_by_id("input-cosd")
        if int(start_input.get_attribute("value")[:4]) < 1993:
            start_input.clear()
            start_input.send_keys(str(start_date))
        else: driver_5.find_element_by_xpath("//span[@id='zoom-all']").click()
        if end_date != None:
            end_input = driver_5.find_element_by_id("input-coed")
            end_input.clear()
            end_input.send_keys(str(end_date))
        driver_5.find_element_by_xpath("//button[@id='download-button']/i[@class='fa fa-download fg-download-btn-gtm']").click()
        element = WebDriverWait(driver_5, 20).until(EC.element_to_be_clickable((By.XPATH, "//a[@id='download-data']")))
        #driver_1.implicitly_wait(3)
        driver_5.find_element_by_xpath("//a[@id='download-data']").click()
```

```python
def clicked():
    global start_date
    global end_date
    if (int(txt.get()) not in range(1800, 2021) or int(txt5.get()) not in range(1800, 2021) or int(txt1.get()) not in range(1, 13) or int(txt2.get()) not in range(1, 32) or int(txt6.get()) not in range(1, 13) or int(txt7.get()) not in range(1, 32)):
        messagebox.showerror('Error', 'Неверный формат даты')
    else:
        start_date = str(txt.get())+"-"+str(txt1.get())+"-"+str(txt2.get())
        lbl4.configure(text=start_date)

        end_date = str(txt5.get())+"-"+str(txt6.get())+"-"+str(txt7.get())
        lbl9.configure(text=end_date)
```

```python
def select_file():
    global file
    file = filedialog.askopenfilename(filetypes = (("Excel files","*.xlsx"), ("Excel files","*.xls")))
```

```python
def download_series_com():
    series_file = pd.read_excel(file, header=None)
    series_list = list(series_file[0])
    download_series_2(series_list = series_list, start_date = start_date, end_date = end_date)
```

```python
window = tkinter.Tk()
window.title("Download series")  
window.geometry('500x250')  

start_lbl = tkinter.Label(window, text="Enter start date") 
start_lbl.grid(column=0, row=0)

lbl = tkinter.Label(window, text="Год:")  
lbl.grid(column=0, row=1)  
txt = tkinter.Entry(window,width=10)  
txt.grid(column=1, row=1) 

lbl1 = tkinter.Label(window, text="Месяц:")  
lbl1.grid(column=0, row=2)  
txt1 = tkinter.Entry(window,width=10)  
txt1.grid(column=1, row=2) 

lbl2 = tkinter.Label(window, text="День:")  
lbl2.grid(column=0, row=3)  
txt2 = tkinter.Entry(window,width=10)  
txt2.grid(column=1, row=3)

lbl3 = tkinter.Label(window, text="Start date:")  
lbl3.grid(column=0, row=4) 

lbl4 = tkinter.Label(window, text="")  
lbl4.grid(column=1, row=4) 

btn = tkinter.Button(window, text="Ok", command=clicked)  
btn.grid(column=5, row=4)



end_lbl = tkinter.Label(window, text="Enter end date") 
end_lbl.grid(column=3, row=0)

lbl5 = tkinter.Label(window, text="Год:")  
lbl5.grid(column=3, row=1)  
txt5 = tkinter.Entry(window,width=10)  
txt5.grid(column=4, row=1) 

lbl6 = tkinter.Label(window, text="Месяц:")  
lbl6.grid(column=3, row=2)  
txt6 = tkinter.Entry(window,width=10)  
txt6.grid(column=4, row=2) 

lbl7 = tkinter.Label(window, text="День:")  
lbl7.grid(column=3, row=3)  
txt7 = tkinter.Entry(window,width=10)  
txt7.grid(column=4, row=3)

lbl8 = tkinter.Label(window, text="End date:")  
lbl8.grid(column=3, row=4) 

lbl9 = tkinter.Label(window, text="")  
lbl9.grid(column=4, row=4) 


txt.focus()

file_lbl = tkinter.Label(window, text="Выберите файл") 
file_lbl.grid(column=0, row=6)

file_lbl1 = tkinter.Label(window, text="со списком названий  ") 
file_lbl1.grid(column=1, row=6)

file_btn = tkinter.Button(window, text="Файл", command=select_file)  
file_btn.grid(column=2, row=6)

dwnld_btn = tkinter.Button(window, text="Скачать", command=download_series_com)  
dwnld_btn.grid(column=0, row=8)


window.mainloop()
```
